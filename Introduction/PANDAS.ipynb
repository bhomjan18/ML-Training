Data Preprocessing in Machine Learning using Pandas

Data processing in required tasks for cleaning the data and making it suitable for a machine learning model which also increase the accuracy and efficiency of a machine learning model

Pandas is an open source Python package that is most widely used for data science/data analysis and ML tasks. It is build on top of another package named Numpy, which provides support for multi-dimensional arrays. Preprocessing is the process of doing a pre-analysis of data, in order to transfer them into a standard and normalized format.

Preprocessing involves thhe following aspects:
* missing values
* data standardization
* data normalization
* data minning



```python
import pandas as pd

data ={
    "cars":["bmw", "honda", "suzuki", "byd"],
    "models": ["2007", "2008", "2015", "2025"],
    "passen": ["4", "6", "6", "5"]
}
data
```

```python
# Convert the data dictionary to dataframe
df = pd.DataFrame(data)
df
```

```python
df.to_csv("/content/drive/MyDrive/Asian college/car_data.csv")
```
