```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("/content/drive/MyDrive/Asian college/Video_Games_Sales_as_at_22_Dec_2016.csv")
df
```

```python
# True represent NaN value
a = df.isnull()
a
```

```python
# Replace NaN in "Name" with the most common name
# cannot change into a cause it's boolean value only
df["Name"].fillna(df["Name"].mode()[0], inplace=True)
df["Genre"].fillna(df["Genre"].mode()[0], inplace=True)
df["Publisher"].fillna(df["Publisher"].mode()[0], inplace=True)
df["Developer"].fillna(df["Developer"].mode()[0], inplace=True)
df["Rating"].fillna(df["Rating"].mode()[0], inplace=True)
df["Year_of_Release"].fillna(df["Year_of_Release"].mean(), inplace=True)
df["Critic_Score"].fillna(df["Critic_Score"].mean(), inplace=True)
df["Critic_Count"].fillna(df["Critic_Count"].mean(), inplace=True)
df["User_Score"].fillna(df["User_Score"].median(), inplace=True)
df["NA_Sales"].fillna(df["NA_Sales"].median(), inplace=True)
df["User_Count"].fillna(df["User_Count"].median(), inplace=True)
df.isnull().sum()
```
