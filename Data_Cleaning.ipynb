# Filtering and Ordering in Pandas
* Read csv file world_population.csv
* Display the list of countries having a rank<= 10
* Create a list of specific country ['Bangladesh', 'Brazil'] and return df which falls under the specific list only
* Apply filter to the Country which has United in its name
* Create a dataframe df2 and give Country as its index
* Apply filter Continent and CCA3 in df2 (need only two columns)
* Apply filter to see details of Zimbabwe entire row
* Sort continent in descending order
* Sort country in ascending order

### Read csv file world_population.csv

```python
df = pd.read_csv("/content/drive/MyDrive/Asian college/ML Project/world_population.csv")
df
```

### Display the list of countries having a rank<= 10

```python
ranked_df = df[df['Rank'] <= 10]
print(ranked_df)
```

### Create a list of specific country ['Bangladesh', 'Brazil'] and return df which falls under the specific list only

```python
specific_countries = ['Bangladesh', 'Brazil']
filtered_df = df[df['Country'].isin(specific_countries)]
filtered_df
```

### Apply filter to the Country which has United in its name

```python
filtered1_df = df[df['Country'].str.contains('United')]
filtered1_df
```

### Create a dataframe df2 and give Country as its index

```python
df2 = df.set_index('Country')
df2
```

### Apply filter Continent and CCA3 in df2 (need only two columns)

```python
df2 = df2[['Continent', 'CCA3']]
df2
```

### Apply filter to see details of Zimbabwe entire row

```python
zimbabwe_details = df[df['Country'] == 'Zimbabwe']
zimbabwe_details
```

### Sort the Continet in descending

```python
df2_sorted = df2.sort_values(by='Continent', ascending = False)
df2_sorted
```

###Sort country in ascending order

```python
df2_sorted = df2.sort_values(by='Country', ascending = True)
df2_sorted
```
