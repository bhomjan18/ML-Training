# Why choose Matplotlib and Seaborn
Both libraries are among the most popular choices for data visualization in Python. Here's why;

Matplotlib: known for its flexibility, Matplotlib allows detailed customization of visual elements, making it perfect for tailored plots.

Seaborn: Built on top of Matplotlib, Seaborn simplifies visualization tasks by offering high-level functions to create aesthically pleasing statistical graphical.

## Visualization with Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Basics of Matplotlib
# line plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, label="Sine wave")
plt.title("Line plot example")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

```python
# Bar chart
categories = ['A', 'B', 'C']
values = [10, 20, 30]
plt.bar(categories, values, color = ['red', 'green', 'blue'])
plt.title("Bar chart example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

## Visualization with seaborn

```python
import seaborn as sns
```

```python
# Histogram and KDE plot:
data = np.random.randn(1000)
sns.histplot(data, kde=True, color = "green")
plt.title("Histogram with KDE")
plt.show()
```

# Heatmap

```python
matrix = np.random.rand(5, 5)
matrix
```

```python
sns.heatmap(matrix, annot=True, cmap="YlOrRd", fmt=".2f")
plt.title("Heatmap Example")
plt.show()
```

# Pairplot Example

```python
sns.set_style("whitegrid")
df = sns.load_dataset("iris")
sns.pairplot(df, hue="species", markers=["o", "s", "D"])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/Asian college/ML Project/Ice Cream Ratings.csv")
df
```

```python
df.set_index("Date")
```

```python
sns.set_style("whitegrid")
sns.scatterplot(data=df, x="Texture Rating", y="Overall Rating")
```

```python
sns.histplot(data=df, x="Overall Rating")
```

```python
sns.boxplot(data=df)
```

```python
df['Flavor Rating'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Ice Cream Flavor Distribution")
plt.show()
```
