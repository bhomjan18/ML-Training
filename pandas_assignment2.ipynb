# Group by and Aggregating

```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Asian college/ML Project/Flavors.csv")
df
```

```python
group_by_frame = df.groupby('Base Flavor')
group_by_frame
```

```python
group_by_frame.mean(numeric_only=True)
```

```python
group_by_frame.count()
```

```python
group_by_frame.sum()
```

```python
df.groupby(['Base Flavor' , 'Liked']).agg({'Flavor Rating' : ['mean' , 'max' , 'count' , 'sum']})
```

```python
# Aggregate using flavor rating and texture rating
df.groupby('Base Flavor').describe()
```

# Merge, Join and Concatenate

```python
import pandas as pd
```

```python
df1 = pd.read_csv("/content/drive/MyDrive/Asian college/ML Project/LOTR.csv")
df1
```

```python
df2 = pd.read_csv("/content/drive/MyDrive/Asian college/ML Project/LOTR 2.csv")
df2
```

```python
pd.concat([df1, df2], ignore_index = True)
```

```python
# Inner Join
df1.merge(df2, how = 'inner' , on = ['FellowshipID', 'FirstName'])
```

```python
# Outer Join
df1.merge(df2, how= 'outer')
```

```python
# Right join
df1.merge(df2, how = 'right')
```

```python
# Left join
df1.merge(df2, how = 'left')
```
